**Problem 1: Unique Words in a Sentence**

Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves



```python
#Given sentance
sentence = input("Original Sentence: ")
sentence = sentence.lower()
words = sentence.split()
unique_words = set(words)
unique_count = len(unique_words)

print("Unique words count:", unique_count)
print("Unique words:", unique_words)
```

**Problem 2: Highest Salary from Employee Data**

Store employee names and their salaries, and determine which employee has the highest salary.


```python
# input given
employees = {
    "Ravi":75000,
    "Anita":68000,
    "Kiran":72000
}

# this helps to find employee with highest salary
highest_employee = max(employees, key = employees.get)
highest_salary = employees[highest_employee]

# print the output
print("Highest Salary:", highest_employee, "-", highest_salary)
```

**Problem 3: Find Maximum and Minimum Values**

Find the maximum and minimum values from a list of numbers.


```python
# given number list
numbers = [45, 22, 89, 10, 66]

max_value = numbers[0] # it stores the maxmium number
min_value = numbers[0] # it stores the minimum number

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List: ", numbers)
print("Max: ", max_value)
print("Min: ", min_value)

```

**Problem 4: Count Products Above a Price Threshold**

Count how many products have a price greater than 1000.


```python
prices = [450,1200,899,1500,300]

count = 0  # this will store how many products are above 1000

# go through each price one by one
for price in prices:
    if price > 1000:   # check if price is greater than 1000
        count = count + 1  # increase count if condition is true

print("Products above 1000:", count)
```

**Problem 5: Calculate Attendance Percentage**

Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ["P","P","A","P","P"]

present_count = 0  # to count number of Present days

# check each day one by one
for day in attendance:
    if day == "P":
        present_count = present_count + 1

total_days = len(attendance)

# calculate percentage
percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

**Problem 7: Count Character Frequency**
    
Count how many times each character appears in a given string.


```python
text = "python"

char_count = {}  # empty dictionary 

for ch in text:
    if ch in char_count:
        char_count[ch] = char_count[ch] + 1
    else:
        char_count[ch] = 1

print(char_count)

```

**Problem 10: Calculate Average Marks**

Calculate the average marks of a student from a list of marks.


```python
#marks are given
marks = [70, 75, 80, 90]

# it add total marks
total = sum(marks)

# it find number of subjects
count = len(marks)

# calculate average
average = total / count

print("Average Marks:", average)
```
